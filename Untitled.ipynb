```python
def plot_fancy_price_action(coins, symbol, start_date, end_date):
    fig = plt.figure()
    dates = []
    high = []
    low = []
    s = 0
    for index, row in coins[start_date:end_date].iterrows():
        if row['symbol'] == symbol:
            if row['date'] == start_date:
                dates.append(row['date'])
            elif row['date'] == start_date:
                dates.append(row['date'])
            else:
                dates.append(row['date'])
            high.append(row['high'])
            low.append(row['low'])

    plt.plot(dates, high, 'r-')
    plt.plot(dates, low, 'g-')
    plt.title('price')
    
    plt.text(20, 550, 'red - highest price on day\ngreen - lowest price on day', fontsize=14)

    plt.grid(True)   # линии вспомогательной сетки
    fig.set_figwidth(20)
    fig.set_figheight(20)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.show()

    
```

3. Накачка и сброс (1 балл)¶


Криптовалютные биржи до сих пор остаются маргинальным местом, эдаким диким западом финансового мира. Как следствие, здесь процветают схемы относительно честного отъема денег. Одна из них - pump'n'dump (накачка и сброс). Она выглядит следуюшим образом. Несколько крупных игроков или много мелких договариваются вместе купить малоизвестную монету с низкой ценой и объемом торгов. Это приводит к мгновенному взлету цены (pump), далее приходят неопытные игроки в надежде успеть заработать на таком росте. В этот момент организаторы схемы начнают все продавать (dump). Весь процесс занимает от нескольких минут до нескольких часов.


Ваша задача найти самый сильный pump'n'dump монеты на заданном промежутке времени. Для этого для каждого дня определим число pnd равное отношению максимальной цены монеты в данный день к максимуму из цен открытия и закрытия в тот же день. Нужно найти день когда pnd был максимален и величину pnd.¶


```python
def find_most_severe_pump_and_dump(coins, symbol, start_date, end_date):
    pnd = 0
    day = ''
    for index, row in coins[start_date:end_date].iterrows():
        if symbol != row['symbol']:
            continue
        mb_pnd = row['high']/max(row['open'], row['close'])
        if mb_pnd > pnd:
            pnd = mb_pnd
            day = index
    return pnd, day
```

```python
find_most_severe_pump_and_dump(coins, symbol='BTC', start_date='2017-06-01', end_date='2018-06-01')
```

Сравните эти значения для разных монет.¶


```python
coin_symbols = set(coins['symbol'])
```

```python
coin_names = []
coin_pnd = []
for name in coin_symbols:
    coin_pnd.append(find_most_severe_pump_and_dump(coins, symbol=name, start_date='2018-05-01', end_date='2018-06-01')[0])
    coin_names.append(name)


dpi = 100
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
mpl.rcParams.update({'font.size': 10})

plt.title('Распределение pnd монет за 6-й месяц 2018 года')

xs = range(len(coin_names))

plt.pie( 
    coin_pnd, 
    radius = 2,
    explode = [0.05] + [0 for _ in range(len(coin_names) - 1)] )
plt.legend(
    bbox_to_anchor = (-0.16, 0.45, 0.25, 0.25),
    loc = 'lower right', labels = coin_names )
plt.show()
```

6. Капитализация как индикатор (1 балл)¶


Многие люди, которые торгуют криптовалютой, любят смотреть на капитализацию. Давайте поймём почему.


Нарисуйте еще два графика. На первом должна быть общая капитализация биткойна (BTC), эфира (ETH), еос (EOS), биткойн кэша (BCH), стеллара (XLM) и лайткойна (LTC). На втором - доли капитализаций этих монет от общей капитализации рынка. При этом используйте данные начиная с 2017-07-01.¶


```python
def plot_coins_capitalizations(coins, symbols, start_date):
    capitalization = dict()
    for index, row in coins[start_date::].iterrows():
        if row['symbol'] in symbols:
            capitalization[row['name']] = 0
    for index, row in coins[start_date::].iterrows():
        if row['symbol'] in symbols:
            capitalization[row['name']] += row['market']
    return capitalization
```

```python
d = plot_coins_capitalizations(
    coins=coins,
    symbols=('BTC', 'ETH', 'EOS', 'BCH', 'XLM', 'LTC'),
    start_date='2017-07-01'
)
```

```python
big_coins = []
capital = []
for key, value in d.items():
    big_coins.append(key)
    capital.append(value)
```

```python
dpi = 100
fig = plt.figure(dpi = dpi, figsize = (512 / dpi, 384 / dpi) )
mpl.rcParams.update({'font.size': 10})

plt.title('Распределение капитализаций монет (%)')

xs = range(len(coin_names))

plt.pie( 
    capital, 
    radius = 2,
    explode = [0.05] + [0 for _ in range(len(big_coins) - 1)] )
plt.legend(
    bbox_to_anchor = (-0.16, 0.45, 0.25, 0.25),
    loc = 'lower right', labels = big_coins )
plt.show()
```

```python
def plot_coins_capitalizations2(coins, symbols, start_date):
    capitalization = dict()
    for index, row in coins[start_date::].iterrows():
        if row['symbol'] in symbols:
            capitalization[row['name']] = 0
        else:
            capitalization['other_market'] = 0
    for index, row in coins[start_date::].iterrows():
        if row['symbol'] in symbols:
            capitalization[row['name']] += row['market']
        else:
            capitalization['other_market'] += row['market']
    return capitalization
```

```python
d = plot_coins_capitalizations2(
    coins=coins,
    symbols=('BTC', 'ETH', 'EOS', 'BCH', 'XLM', 'LTC'),
    start_date='2017-07-01'
)
```
